```python
# agregar categoria de edad
# nio <12, adolescente 12 a 17, adulto 18 a 55, adulto mayor > 55
# Pasos: 1) Crear un df con datos "no nulos"
# 2) Crear una lista con las categorias de edad
# 3) Crear una funcion que recorra el df y agregue la categoria de edad
# 4) Imprimir el df con la categoria de edad agregada
```

```python
print(df_titanic.columns.tolist())
```

```python
# Crea un DF con los valores a utilizar
df_1 = df_titanic[['Name', 'Age']]
```

```python
# Crear un DF con edad valida
df_2 = df_1[df_1['Age'].notna()]
```

```python
# Agregar una nueva columna llamada "Category"
df_2['Category'] = df_2['Age']
```

```python
df_2.info()
```

```python
# Construir la funcion en python para categorias
def category_f(age):
    if (age >0) and (age <= 1):
        return "baby"
    elif (age > 1) and (age < 12):
        return "child"
    elif (age >= 12) and  (age < 18):
        return "teenager"
    elif (age >= 18) and (age < 55):
        return "adult"
    else:
        return "senior"
```

```python
category_f(1)
```

```python
# Aplicamos la funcion .apply a la columna (atributo) Category

df_2['Category'].apply(category_f)
```

```python
# Modificacion del DF para incluir la categoria con sus valores
df_2['category'] = df_2['Category'].apply(category_f)
```
